Task 1: Prediction using Supervised ML 


```python
#STEP 1:  Importing necessary libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt  
```

```python
#STEP 2: Importing the dataset
data = "http://bit.ly/w-data"
data_set = pd.read_csv(data)
print("First 10 elements of the given dataset:")
print(data_set.head(10))
```

```python
# STEP 3: Separating dependent and independent variables
x = data_set.iloc[:, 0:1].values  # Hours studied
y = data_set.iloc[:, 1:2].values  # Scores
```

```python
# STEP 4: Visualizing the data with a scatter plot
plt.scatter(x, y, color='blue')
plt.title("Hours Studied vs Percentage Score")
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
# STEP 5: Splitting data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
# Visualizing the training data
plt.scatter(x_train, y_train, color='green')
plt.title("Training Data (Hours Studied vs Percentage Score)")
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
# STEP 6: Training the Linear Regression model
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
lg.fit(x_train, y_train)
print("Training complete")
```

```python
# STEP 7: Making Predictions
y_predict = lg.predict(x_test)
print("Predicted scores on test data:")
print(y_predict)
```

```python
# STEP 8: Checking the accuracy of the model with a plot
plt.scatter(x_train, y_train, color='green')
plt.plot(x_test, y_predict, color="red") 
plt.title("Regression Line (Test Data)")
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
# STEP 9: Testing data - In hours
print("Test data (hours studied):")
print(x_test)
```

```python
# STEP 10: Predicting the score for 9.25 hours/day
predicted_score = lg.predict([[9.25]])
print(f"Predicted score for 9.25 hours/day: {predicted_score[0][0]:.2f}")
```

```python
# STEP 11: Evaluating the model
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_predict)
print(f"R-squared value of the model: {r2:.2f}")
```
